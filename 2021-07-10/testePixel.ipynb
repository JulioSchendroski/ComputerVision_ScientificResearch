```python
import cv2 as cv
import numpy as np
from matplotlib import image, pyplot as plt
```

```python
img = cv.imread("7.jpeg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
resized = cv.resize(gray, (600,400))
_,bin = cv.threshold(resized,30,255,cv.THRESH_BINARY)

plt.imshow(bin, cmap='gray')
plt.show()
```
